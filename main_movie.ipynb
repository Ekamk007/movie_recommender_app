# Data Preprocessing

```python
def convert(obj):
    L = []

    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

```python
def convert3(obj):
    L = []
    counter = 0

    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break

    return L
```

```python
def fetch_director(obj):
    L = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert3)
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])

movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])

movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])

movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + 3*movies['genres'] + 2*movies['keywords'] + 1*movies['cast'] + 2*movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

# **Stemming**

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# **Sentence Transformer**

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
#-----------------------------
# 1. Load embedding model
# -----------------------------
model = SentenceTransformer('all-MiniLM-L6-v2')

# -----------------------------
# 2. Generate embeddings
# -----------------------------
embeddings = model.encode(new_df['tags'].tolist(), show_progress_bar=True)

# -----------------------------
# 3. Cosine similarity matrix
# -----------------------------
similarity = cosine_similarity(embeddings)

# -----------------------------
# 4. Index mapping
# -----------------------------
indices = pd.Series(new_df.index, index=new_df['title']).drop_duplicates()

# -----------------------------
# 5. Recommendation function
# -----------------------------
def recommend(movie_title, top_n=5):
    idx = indices[movie_title]

    sim_scores = list(enumerate(similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:top_n+1]

    return [new_df.iloc[i[0]]['title'] for i in sim_scores]

# -----------------------------
# 6. Example
# -----------------------------
print(recommend("Avatar"))
```

```python
final_sim = (
    0.1 * cv_similarity +
    0.3 * tf_idf_sim +
    0.6 * similarity
)
```

```python
def recommend(movie):
    idx = new_df[new_df['title'] == movie].index[0]

    scores = list(enumerate(final_sim[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:6]

    for i in scores:
        print(new_df.iloc[i[0]].title)
```

```python
def fetch_poster(movie_id):

    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}&language=en-US"

    data = requests.get(url)

    data = data.json()

    poster_path = data.get('poster_path')

    if poster_path:
        full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    else:
        # Return a placeholder image if poster_path is not available
        full_path = "https://via.placeholder.com/500x750?text=No+Poster+Available"

    return full_path
```

```python
def fetch_reviews(movie_id):

    url = f"https://api.themoviedb.org/3/movie/{movie_id}/reviews?api_key={API_KEY}&language=en-US"

    data = requests.get(url)

    data = data.json()

    reviews = []

    for review in data.get('results', []):
        reviews.append(review['content'])

    return reviews
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
```

```python
classifier("This movie was absolutely amazing")
```

```python
def sentiment_score(reviews):

    if len(reviews) == 0:
        return 0

    scores = []

    for review in reviews:

        result = classifier(review[:512])[0]

        if result['label'] == 'POSITIVE':
            scores.append(result['score'])

        else:
            scores.append(1 - result['score'])

    return sum(scores) / len(scores)
```

```python
reviews = fetch_reviews(19995)

score = sentiment_score(reviews)

score
```

```python
def hybrid_recommend(movie):

    movie_index = new_df[new_df['title'] == movie].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x:x[1]
    )[1:11]

    final_movies = []

    for i in movies_list:

        idx = i[0]

        similarity_score = i[1]

        movie_id = new_df.iloc[idx].movie_id

        title = new_df.iloc[idx].title

        poster = fetch_poster(movie_id)

        reviews = fetch_reviews(movie_id)

        sentiment = sentiment_score(reviews)

        final_score = (0.8 * similarity_score) + (0.2 * sentiment)

        final_movies.append({
            'title': title,
            'poster': poster,
            'similarity': similarity_score,
            'sentiment': sentiment,
            'final_score': final_score
        })

    final_movies = sorted(
        final_movies,
        reverse=True,
        key=lambda x:x['final_score']
    )

    return final_movies[:5]
```

```python
results = hybrid_recommend('Atonement')
```

```python
for movie in results:

    print(movie['title'])

    print("Similarity:", movie['similarity'])

    print("Sentiment:", movie['sentiment'])

    print("Final Score:", movie['final_score'])

    print()
```
